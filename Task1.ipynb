## Pairwise Euclidean Distance

````python
def pairwise_euclidian_matrix(inertial_signal):
    euclidian_similarity_matrix = np.zeros((1500,1500))
    for i in range(1500):
        for j in range(i+1, 1500):
            distance = np.linalg.norm(inertial_signal[i] - inertial_signal[j])
            euclidian_similarity_matrix[i][j] = distance
            euclidian_similarity_matrix[j][i] = distance
    return euclidian_similarity_matrix 
````

## Pairwise Dynamic Time Warping (DTW) distance

The below is implementation of DTW algorithm from scratch. However it is very slow and takes lot of time to run. Hence we use dtaidistance library to computute DTW distances between sequences and compute the pairwise DTW matrix.

```python
def dtw_distance(seq1_x, seq1_y, seq1_z, seq2_x, seq2_y, seq2_z): 
    n = len(seq1_x)
    m = len(seq2_x)
    dtw = np.full((n+1, m+1), np.inf)
    dtw[0, 0] = 0
    
    for i in range(1, n+1):
        for j in range(1, m+1):
            cost = np.sum(abs(seq1_x[i-1] - seq2_x[j-1]) + abs(seq1_y[i-1] - seq2_y[j-1]) + abs(seq1_z[i-1] - seq2_z[j-1]))
            dtw[i, j] = cost + min(dtw[i-1, j], dtw[i, j-1], dtw[i-1, j-1])
    
    return dtw[n, m]

def calculate_dtw_matrix(signal_x, signal_y, signal_z):
    num_signals = len(signal_x)
    dtw_matrix = np.zeros((num_signals, num_signals))
    
    for i in range(num_signals):
        for j in range(i+1, num_signals):
            distance = dtw_distance(signal_x[i], signal_y[i], signal_z[i], signal_x[j], signal_y[j], signal_z[j])
            dtw_matrix[i, j] = distance
            dtw_matrix[j, i] = distance  # Symmetry
    
    return dtw_matrix

## Diffusion Maps and K-Means Clustering

````python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
````

````python
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import adjusted_rand_score, silhouette_score
````

````python
def plot_diffusion_embeddings(embedding, labels, dim):

    activity_labels = {1: 'WALKING', 2: 'WALKING_UPSTAIRS', 3: 'WALKING_DOWNSTAIRS', 
                       4: 'SITTING', 5: 'STANDING', 6: 'LAYING'}

    if dim == 2:
        print("\nPLOT WITH 2 EMBEDDING VECTORS")  
        plt.figure(figsize=(6, 6))
        # Map numeric labels to activity names
        str_labels = [activity_labels[label] for label in labels]
        sns.scatterplot(x=embedding[:, 0], y=embedding[:, 1], hue=str_labels, 
                        palette="tab10", alpha=0.8, edgecolor="black")
        plt.xlabel("Diffusion Coordinate 1")
        plt.ylabel("Diffusion Coordinate 2")
        plt.title("Diffusion Map Embedding (2D)")
        plt.legend(title="Activity Label", bbox_to_anchor=(1.05, 1), loc='upper left')
        plt.grid(True)
        plt.show()


    elif dim == 3:
        print("\nPLOT WITH 3 EMBEDDING VECTORS")  
        fig = plt.figure(figsize=(7, 6))  # Slightly larger figure
        ax = fig.add_subplot(111, projection='3d')
        scatter = ax.scatter(embedding[:, 0], embedding[:, 1], embedding[:, 2], 
                            c=labels, cmap="tab10", alpha=0.8, edgecolors="black")

        ax.set_xlabel("Diffusion Coordinate 1")
        ax.set_ylabel("Diffusion Coordinate 2")
        ax.set_zlabel("Diffusion Coordinate 3")  # Increased padding
        ax.set_title("Diffusion Map Embedding (3D)")

        # Add colorbar with adjusted position
        cbar = fig.colorbar(scatter, ax=ax, shrink=0.7, pad=0.1)  # Added pad parameter
        cbar.set_label("Activity Label")
        
        # Set ticks and labels
        cbar.set_ticks(list(activity_labels.keys()))
        cbar.set_ticklabels(list(activity_labels.values()))

        plt.tight_layout()  # Automatically adjust subplot params
        plt.show()

    else:
        raise ValueError("dim must be 2 or 3")
````

````python
def cluster(embeddings):
    n_clusters = 6
    true_labels = y_train[:1500].squeeze()
    kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
    kmeans_labels = kmeans.fit_predict(embeddings)
    
    ari_kmeans = adjusted_rand_score(true_labels, kmeans_labels)
    silhouette_kmeans = silhouette_score(embeddings, kmeans_labels)

    # PLotting the KMeans Clusters
    print("\nPLOTTING KMEANS CLUSTERS")
    plot_diffusion_embeddings(embeddings, kmeans_labels, dim=3)

    # Print results
    print("="*50)
    print("K-MEANS CLUSTERING RESULTS:")
    print(f"ARI: {ari_kmeans:.4f}, Silhouette Score: {silhouette_kmeans:.4f}")
    print("="*50)
````

````python
def diffusion_cluster(dtw_matrix,n):
    epsilon = np.median(dtw_matrix.ravel())**n
    L = np.exp(-dtw_matrix**2 / epsilon) 
    D = np.diag(L.sum(axis=1))
    L_alpha = np.linalg.inv(D) @ L @ np.linalg.inv(D) 
    D_alpha = np.diag(L_alpha.sum(axis=1))  
    M = np.linalg.inv(D_alpha) @ L_alpha    

    # Compute eigenvectors
    eigenvalues, eigenvectors = np.linalg.eig(M) # Since M is symmetric, M and M.T have the same eigenvalues and eigenvectors

    eigenvalues = eigenvalues.real
    eigenvectors = eigenvectors.real    

    sorted_idx = np.argsort(eigenvalues)[::-1]  # Descending order
    eigenvalues_sorted = eigenvalues[sorted_idx]
    eigenvectors_sorted = eigenvectors[:, sorted_idx]
    print("\nEIGEN VALUES:")
    print(eigenvalues_sorted)
    
    # Extract top 3 eigenvectors (excluding the constant eigenvector)
    reduced_data = eigenvectors_sorted[:, 1:4] * eigenvalues_sorted[1:4]
    print("\nEMBEDDINGS:")
    print(reduced_data)

    # 2D Representation of Data
    plot_diffusion_embeddings(reduced_data, y_train[:1500].squeeze(), dim=2)

    # 3D Representation of Data
    plot_diffusion_embeddings(reduced_data, y_train[:1500].squeeze(), dim=3)

    # Clustering
    cluster(reduced_data)
````

## Diffusion Clustering with $ \epsilon $ = MEDIAN OF DISTANCE MATRIX ** 1

````python
# Epsilon value affects the clustering results.
# Let epsilon = (median of the pairwise distance matrix) ^ n_factor

print(f"DIFFUSION CLUSTERING WITH EPSILON = MEDIAN OF DISTANCE MATRIX ** 1")
diffusion_cluster(dtw_matrix, 1)
print("\n"+"="*100)
````

## Diffusion Clustering with $ \epsilon $ = MEDIAN OF DISTANCE MATRIX ** 2

````python
print(f"DIFFUSION CLUSTERING WITH EPSILON = MEDIAN OF DISTANCE MATRIX ** 2")
diffusion_cluster(dtw_matrix, 2)
print("\n"+"="*100)
````

## Diffusion Clustering with $ \epsilon $ = MEDIAN OF DISTANCE MATRIX ** 3

````python
print(f"DIFFUSION CLUSTERING WITH EPSILON = MEDIAN OF DISTANCE MATRIX ** 3")
diffusion_cluster(dtw_matrix, 3)
print("\n"+"="*100)
````

### RESULTS

$ \epsilon $ = (median of pairwise distance matrix)$ ^n $

- For $ n $ = 1: We get ARI = 0.0007 and Silhouette Score = 0.9844

- For $ n $ = 2: We get ARI = 0.4286 and Silhouette Score = 0.6749

- For $ n $ = 3: We get ARI = 0.4468 and Silhouette Score = 0.4565

$ \therefore $ We get best results when $ \epsilon $ = (median of pairwise distance matrix)$ ^2 $

## Observations and Discussions

We observe that we get better clustering results using diffusion maps, with dynamic time warping as distance metric when compared to PCA or TSNE. Diffusion maps work better because time-series data lie on low-dimensional manifolds embedded in high-dimensional spaces which are preserved. Also, it uses local similarity(dynamic time warping) to uncover complex nonlinear data manifold.

PCA projects the points on to a line. This results in loss of non-linearity.

t-SNE is good for preserving the local non-linear structure, however it can not maintain the global non-linear structure.

Raw features, even though contain all the information, do not form good clusters because of lot of redundent information and noise.
